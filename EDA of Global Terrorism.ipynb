Treating the null values

```python

# Calculate the total number of rows
total_rows = len(globalterrorism_df)

# Calculate the number of missing values in each column
missing_data = globalterrorism_df.isnull().sum()

# Calculate the percentage of missing data for each column
percentage_missing = (missing_data / total_rows) * 100

# Create a DataFrame to display the results
missing_data_summary = pd.DataFrame({
    'Column': missing_data.index,
    'Missing Values': missing_data.values,
    'Percentage Missing': percentage_missing.values
})

# Sort the DataFrame by the percentage of missing data in descending order
missing_data_summary = missing_data_summary.sort_values(by='Percentage Missing', ascending=False)

# Display the summary
print(missing_data_summary)
```

```python
# Columns to fill with 'Unknown'
columns_to_fill = ['State', 'City', 'Motive', 'Summary', 'Target']

# Iterate through columns and fill with 'Unknown'
for column in columns_to_fill:
    globalterrorism_df[column] = globalterrorism_df[column].fillna('Unknown')

# Columns to impute with median
columns_to_impute_with_median = ['Wounded', 'Died']

# Iterate through columns and impute with median
for column in columns_to_impute_with_median:
    median_value = globalterrorism_df[column].median()
    globalterrorism_df[column].fillna(median_value, inplace=True)
```

Feature Relationship

```python
# Group the data by 'Region' and count the number of incidents
region_counts = globalterrorism_df['Region'].value_counts().reset_index()
region_counts.columns = ['Region', 'Incident Count']

# Sort the regions by incident count in descending order
region_counts = region_counts.sort_values(by='Incident Count', ascending=False)

# Set a threshold for identifying high-risk regions (e.g., regions with a high incident count)
threshold = 500  # Adjust this threshold as needed

# Filter regions that exceed the threshold
high_risk_regions = region_counts[region_counts['Incident Count'] >= threshold]

# Plot the high-risk regions using a bar plot
sns.set_palette('viridis')
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Region', y='Incident Count', data=high_risk_regions)
plt.title('High-Risk Regions Based on Incident Count', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Incident Count', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Print the list of high-risk regions
print("High-Risk Regions:")
print(high_risk_regions)
```

Terrorist attacks that led to death

```python
Died_by_Attacks = globalterrorism_df.loc[:,'Died']
print('Number of deaths as a result of the terrorist attack:', int(sum(Died_by_Attacks)))
```

```python
Attacks_Died = globalterrorism_df.loc[:,'Attack_type']
Died_type_Attacks = pd.concat([Attacks_Died, Died_by_Attacks], axis=1)
```

```python
Deathtype = Died_type_Attacks.pivot_table(columns='Attack_type', values='Died', aggfunc='sum')
Deathtype
```

```python

# Extract labels and transpose the data
attack_types = Deathtype.columns.tolist()
transposed_data = Deathtype.T
attack_values = transposed_data.values.flatten()
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(aspect="equal"))
wedges, texts, autotexts = plt.pie(attack_values, labels=attack_types, autopct='%.1f%%', pctdistance=0.85, colors=colors,
                                    startangle=90, shadow=True, wedgeprops={'linewidth': 2, 'edgecolor': 'w'})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
plt.title('Terrorist attacks that led to death', fontsize=16)
plt.legend(attack_types, title='Attack Types', loc='upper left', bbox_to_anchor=(1.2, 1))
for text, autotext in zip(texts, autotexts):
    text.set(size=12, weight='bold')
    autotext.set(size=12, weight='bold')
ax.axis('equal')
plt.show()
```

After a careful analysis displayed above, the following key findings were observed:

Highest Activity Year: In 2014, terrorist activities reached their peak, indicating a concerning escalation.

Peak Activity in May: May emerges as the month with the highest frequency of terrorist attacks,
suggesting a seasonal pattern that may warrant further investigation.

Day of the Month Trends: A noteworthy trend is that terrorist attacks peak on the 15th day of the month
but subsequently decrease towards the end (last 2-3 days). This cyclic behaviour could have significant implications.

High-Risk Regions: The Middle East & North Africa stands out as the region with the highest potential risk for terrorist attacks,
followed closely by South Asia. Understanding the factors contributing to this trend is vital for security measures.

Preferred Attack Methods: The predominant methods of terrorist attacks are Bombing/Explosion, followed by Armed Assault.
Recognising these patterns can help in risk assessment and prevention strategies.

Primary Targets: Private citizens and property appear to be the primary targets of terrorist groups, emphasizing the need for measures to safeguard civilians and assets.

Success Rate: The success rate of terrorist attacks over the past 47 years is alarmingly high at 88.96%. A closer look into the factors contributing to this high rate is essential for counter-terrorism efforts.

Weapon of Choice: Explosives are the most frequently used weapon type in terrorist attacks. This knowledge can inform security measures and resource allocation.

High-Risk Country and City: Iraq emerges as the most attacked country, and the top city attacked remains unknown. Efforts to address terrorism must consider these hotspots as priorities.
